```python
import numpy as np
import soundfile as sf
import matplotlib.pyplot as plt
from matplotlib import cm

# Load audio file (replace with your audio file path)
audio_file = '/standard/UVA-DSA/NIST EMS Project Data/CognitiveEMS_Datasets/North_Garden/Sep_2024/Raw/05-09-2024/debrah/cardiac_arrest/2/audio/GX010334-encoded.mp3'
audio_data, samplerate = sf.read(audio_file)

# Downsample the data aggressively for faster plotting
time = np.linspace(0, len(audio_data) / samplerate, num=len(audio_data))

# Create a figure with a dark background
plt.figure(figsize=(12, 6), facecolor='white')

# Plot the downsampled waveform with a gradient color effect
plt.plot(time, audio_data, color='blue', linewidth=0.8)

# Add labels and title with modern fonts and colors
plt.title('Audio Waveform', fontsize=20, color='black', pad=20)
plt.xlabel('Time (s)', fontsize=14, color='black')
plt.ylabel('Amplitude', fontsize=14, color='black')

# Change the color of ticks
plt.tick_params(axis='x', colors='black')
plt.tick_params(axis='y', colors='black')

# remove axis lines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
# Set dark gridlines
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.5, color='gray')

# Set the background color of the plot
plt.gca().set_facecolor('white')

# Improve layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the audio file
file, sr = librosa.load(audio_file)

# Create the plot
plt.figure(figsize=(14, 5))

# Plot the waveform with styling
librosa.display.waveshow(file, sr=sr, alpha=0.8)

# Add a title and labels
plt.title('Waveform of Audio', fontsize=15)
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)

# Customize grid and remove top/right borders
# plt.grid(True)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# Show the plot
plt.show()
```

```python
# Create the Mel spectrogram
S = librosa.feature.melspectrogram(y=file, sr=sr, n_mels=128, fmax=8000)
S_dB = librosa.power_to_db(S, ref=np.max)

# Create a new figure for the spectrogram
plt.figure(figsize=(14, 5))

# Plot the Mel spectrogram
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', fmax=8000, cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram', fontsize=15)
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Frequency (Hz)', fontsize=12)

# Show the spectrogram plot
plt.show()
```

```python
import json
import os
```

```python

via_raw_annotations = '/standard/UVA-DSA/NIST EMS Project Data/EgoExoEMS_CVPR2025/Dataset/Final/annotations/via_annotation_paths.txt'
gopro_original_paths = '/standard/UVA-DSA/NIST EMS Project Data/CognitiveEMS_Datasets/North_Garden/Final/gopro_paths.txt'
gopro_final_paths = '/standard/UVA-DSA/NIST EMS Project Data/EgoExoEMS_CVPR2025/Dataset/Final/gopro_paths.txt'
```

```python
# read text file and load all paths
via_json_paths = None
with open(via_raw_annotations, 'r') as f:
    via_json_paths = f.readlines()
    via_json_paths = [x.strip() for x in via_json_paths]

print(via_json_paths)
    
```

```python
# Load GoPro original paths
gopro_original = None
with open(gopro_original_paths, 'r') as f:
    gopro_original = f.readlines()
    gopro_original = [x.strip() for x in gopro_original]

print("Original Paths:")
print(gopro_original)

# Load GoPro final paths
gopro_final = None
with open(gopro_final_paths, 'r') as f:
    gopro_final = f.readlines()
    gopro_final = [x.strip() for x in gopro_final]

print("Final Paths:")
print(gopro_final)

# Create a mapping from original file names to original paths
original_mapping = {path.split('/')[-1]: path for path in gopro_original}

# Create a mapping from final paths to original paths
final_to_original_mapping = {}
for final_path in gopro_final:
    file_name = final_path.split('/')[-1]
    if file_name in original_mapping:
        final_to_original_mapping[final_path] = original_mapping[file_name]


# sort the mapping using original date in original path
final_to_original_mapping = dict(sorted(final_to_original_mapping.items(), key=lambda x: x[1]))

# Print the mapping
print("Final to Original Mapping:")
for final, original in final_to_original_mapping.items():
    print(f"{final} -> {original}")

# save the mapping to a text file
mapping_file = '/standard/UVA-DSA/NIST EMS Project Data/EgoExoEMS_CVPR2025/Dataset/Final/gopro_final_to_original_mapping.txt'
with open(mapping_file, 'w') as f:
    for final, original in final_to_original_mapping.items():
        f.write(f"{final} -> {original}\n")
```

## VIA annotation splitter

```python
# load via annotation that has multiple videos
via_multiple_json = '/standard/UVA-DSA/NIST EMS Project Data/EgoExoEMS_CVPR2025/Dataset/Final/ng8/cardiac_arrest/8-10_annotation.json'
output_dir = '/standard/UVA-DSA/NIST EMS Project Data/EgoExoEMS_CVPR2025/Dataset/Final/ng8/cardiac_arrest/'


# Load the original annotation file
file_path = via_multiple_json
with open(file_path, 'r') as f:
    annotation_data = json.load(f)

# Create separate annotation files for each video
os.makedirs(output_dir, exist_ok=True)

# Iterate over videos and create a new file for each
for view_id, view_data in annotation_data['view'].items():
    video_fid = view_data['fid_list'][0]
    video_annotations = {
        "project": annotation_data['project'],
        "config": annotation_data['config'],
        "attribute": annotation_data['attribute'],
        "file": {video_fid: annotation_data['file'][video_fid]},
        "metadata": {k: v for k, v in annotation_data['metadata'].items() if v['vid'] == video_fid},
        "view": {view_id: view_data}
    }
    # get file name
    file_name = annotation_data['file'][video_fid]['fname'].split('/')[-1].split('.')[0]
    print(f"Saving annotation for {file_name}")
    
    # # Save the separate annotation file for the current video
    output_file = os.path.join(output_dir, f"annotation_trial_{file_name}.json")
    with open(output_file, 'w') as out_f:
        json.dump(video_annotations, out_f, indent=2)

print(f"Separate annotation files have been created in: {output_dir}")
```

```python
import h5py
import matplotlib.pyplot as plt

# Open the HDF5 file
with h5py.File("../../../TestData/anonymous/ng3_t0_ks4/depth_ir_data.hdf5", "r") as f:
    # Load a depth frame
    depth_data = f["depth/frame_39"][:]
    plt.imshow(depth_data, cmap="gray")
    plt.title("Depth Frame")
    plt.colorbar()
    plt.show()

    # Load an IR frame
    ir_data = f["ir/frame_39"][:]
    plt.imshow(ir_data, cmap="gray")
    plt.title("IR Frame")
    plt.colorbar()
    plt.show()
```

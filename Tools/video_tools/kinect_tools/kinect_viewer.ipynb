```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


path = '/standard/UVA-DSA/NIST EMS Project Data/CognitiveEMS_Datasets/North_Garden/Sep_2024/Raw/05-09-2024/debrah/cardiac_arrest/3/Kinect/ir/ir0123.png'

# Load the depth image (assumed to be in 16-bit PNG format)
depth_image = cv2.imread('depth_image.png', cv2.IMREAD_UNCHANGED)

# Check the shape of the image (height, width)
print("Depth image shape:", depth_image.shape)

# Print depth value at a specific pixel (for example, at (100, 150))
x, y = 100, 150
depth_value = depth_image[y, x]
print(f"Depth at pixel ({x}, {y}): {depth_value} mm")

# Alternatively, print all depth values (optional, might be large)
# print(depth_image)
```

```python
# Normalize the depth values for visualization (0 to 255 range for heatmap)
depth_normalized = cv2.normalize(depth_image, None, 0, 255, cv2.NORM_MINMAX)
depth_normalized = np.uint8(depth_normalized)  # Convert to 8-bit for display

# Create a heatmap using Matplotlib's 'hot' colormap
plt.figure(figsize=(10, 8))
plt.imshow(depth_normalized, cmap='hot')
plt.colorbar(label='Depth (normalized)')
plt.title('Depth Heatmap')
plt.show()
```
